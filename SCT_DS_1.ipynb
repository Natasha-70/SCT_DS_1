```python
# Data from the image
age_groups = ['0-20 Years', '21-64 Years', '65+ Years']
population_millions = [512, 807, 98]  # In millions
percentages = [36.1, 57.0, 6.9]
# Colors (as per the image: Yellow, Blue, Pink)
colors = ['#ffeb3b', '#2196f3', '#e91e63']
# Plotting the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(age_groups, population_millions, color=colors)
# Add value labels on top of each bar
for bar, percentage in zip(bars, percentages):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 10,f'{height}M\n({percentage}%)', ha='center', fontsize=12)
# Chart title and labels
plt.title("India's Population Distribution by Age Group (2022)", fontsize=14)
plt.xlabel("Age Groups")
plt.ylabel("Population (in millions)")
plt.ylim(0, 900)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Show the plot
plt.tight_layout()
plt.show()
```
